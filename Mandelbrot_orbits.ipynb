Plot the iterations of 0 under 


$$P_c: \mathbb{C} \mapsto \mathbb{C}$$
$$P_c: z \mapsto z^2 + c$$

and lay them out in $\mathbb{C}$

```python
%pylab inline
from matplotlib.backends.backend_pdf import PdfPages
```

```python
def P(c):
    z = 0
    count = 0
    iters = []
    while (np.linalg.norm(z) < 2) and (count < 500):
        iters.append(z)
        try:
            z = z**2 + c
        except:
            z = np.complex(3,3)
        count += 1
    return count, iters

def plot_iters(iters,ax=None):
    a = [v.real for v in iters[1:]]
    b = [v.imag for v in iters[1:]]
    if ax is None:
        plt.plot(a,b)
    else:
        ax.plot(a,b)

count, iters = P(np.complex(0.31111,0.41111))
plot_iters(iters)
```

```python
res = 50
rs = [-1.5, 0.5]
bs = [0.0001, 1]

params = tuple(rs + bs + [res])

with PdfPages('orbits_{}-{}_{}-{}_{}.pdf'.format(*params)) as pdf:
    fig, ax = plt.subplots(res, res,figsize=(50, 50))
    for i,r in enumerate(np.linspace(rs[0], rs[1], res)):
        print(i, end=' ')
        for j,b in enumerate(np.linspace(bs[1], bs[0], res)):
            count, iters = P(np.complex(r,b))
            plot_iters(iters, ax=ax[j,i])
            ax[j,i].axis('off')
            #ax[j,i].set_title('{}+{}i'.format(r,b))
    pdf.savefig(fig, bbox_inches = 'tight', pad_inches = 0.5)
    plt.close()   
```
